```python
from langchain_community.document_loaders import PyPDFLoader
loader = PyPDFLoader("ramayana_nlp.pdf")
docs = loader.load()
docs
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=20)
text_splitter.split_documents(docs)[:5]
```

```python
documents=text_splitter.split_documents(docs)
documents
```

```python
len(documents)
```

```python
pip install faiss-cpu
```

```python
from langchain_community.embeddings import OllamaEmbeddings
from langchain_community.vectorstores import FAISS

db=FAISS.from_documents(documents,OllamaEmbeddings())
```

```python
db
```

```python
query="Vishwamitra, Rama and Lakshmana reached Mithila in time for Sita's swamyamvar"
result=db.similarity_search(query)
result[0].page_content
```

```python
embedding_vector = OllamaEmbeddings().embed_query(query)
docs = db.similarity_search_by_vector(embedding_vector)
print(docs[0].page_content)
```

```python
from langchain_community.llms import Ollama
## Load Ollama LAMA2 LLM model
llm=Ollama(model="llama2")
llm
```

```python
## Design ChatPrompt Template
from langchain_core.prompts import ChatPromptTemplate
prompt = ChatPromptTemplate.from_template("""
Answer the following question based only on the provided context. 
Think step by step before providing a detailed answer. 
<context>
{context}
</context>
Question: {input}""")
```

```python
## Chain Introduction
## Create Stuff Docment Chain

from langchain.chains.combine_documents import create_stuff_documents_chain

document_chain=create_stuff_documents_chain(llm,prompt)
```

```python
"""
Retrievers: A retriever is an interface that returns documents given
 an unstructured query. It is more general than a vector store.
 A retriever does not need to be able to store documents, only to 
 return (or retrieve) them. Vector stores can be used as the backbone
 of a retriever, but there are other types of retrievers as well. 
 https://python.langchain.com/docs/modules/data_connection/retrievers/   
"""

retriever=db.as_retriever()
retriever
```

```python
"""
Retrieval chain:This chain takes in a user inquiry, which is then
passed to the retriever to fetch relevant documents. Those documents 
(and original inputs) are then passed to an LLM to generate a response
https://python.langchain.com/docs/modules/chains/
"""
from langchain.chains import create_retrieval_chain
retrieval_chain=create_retrieval_chain(retriever,document_chain)
```

```python
response=retrieval_chain.invoke({"input":"Rama"})
print(response['answer'])
```

```python
response=retrieval_chain.invoke({"input":"Who are the main characters in Ramayana?"})
print(response['answer'])
```

```python
response=retrieval_chain.invoke({"input":"Name the four sons of king dasaratha"})
print(response['answer'])
```

```python
response=retrieval_chain.invoke({"input":"How long was Rama's exile"})
print(response['answer'])
```
